```python
phi = (1+sqrt(5))/2
icosaColor = ['black','red','orange','darkkhaki','lawngreen','green','darkseagreen','fuchsia','pink','indigo','cadetblue','cornflowerblue','aquamarine']
icosaVertices = [(0,0,0),(1,phi,0),(-1,phi,0),(0,1,phi),(-1,-phi,0),(0,-1,phi),(-phi,0,1),(0,1,-phi),(0,-1,-phi),(phi,0,1),(1,-phi,0),(phi,0,-1),(-phi,0,-1)]
centered = {'0' : [0,[1,2,3],[4,5,6],[9,10,11],[7,8,12]],
            '1' : [1,[0,2,3],[4,8,9],[6,7,11],[5,10,12]],
            '2' : [2,[0,1,3],[4,11,12],[6,8,10],[5,7,9]],
            '3' : [3,[0,1,2],[4,7,10],[5,8,11],[6,9,12]],
            '4' : [4,[0,5,6],[1,8,9],[3,7,10],[2,11,12]],
            '5' : [5,[0,4,6],[1,10,12],[3,8,11],[2,7,9]],
            '6' : [6,[0,4,5],[1,7,11],[2,8,10],[3,9,12]],
            '7' : [7,[0,8,12],[1,6,11],[3,4,10],[2,5,9]],
            '8' : [8,[0,7,12],[1,4,9],[3,5,11],[2,6,10]],
            '9' : [9,[0,10,11],[1,4,8],[2,5,7],[3,6,12]],
            '10': [10,[0,9,11],[1,5,12],[3,4,7],[2,6,8]],
            '11': [11,[0,9,10],[1,6,7],[3,5,8],[2,4,12]],
            '12': [12,[0,7,8],[1,5,10],[2,4,11],[3,6,9]]}

class icosaPuzzle:
    def __init__(self):
        self.center = 0
        return
    def icosaGenTotal(self):
        Lines = point((0,0,0), size = 10, color = 'white')
        for k in range(13):
            temp = centered[str(k)]
            for i in range(1,5):
                for j in range(3):
                    Lines += line([icosaVertices[temp[0]],icosaVertices[temp[i][j]]], color = 'black')
                    Lines += line([icosaVertices[temp[i][(j-1)%3]],icosaVertices[temp[i][j]]], color = 'black')
        for i in range(13):
            Lines += point(icosaVertices[i], color = icosaColor[i],size = 40)
        show(Lines,axes = False)
        return
    def icosaGenLocal(self):
        temp = centered[str(self.center)]
        Lines = point((0,0,0), color = 'white')
        for i in range(1,5):
            for j in range(3):
                Lines += line([icosaVertices[temp[0]],icosaVertices[temp[i][j]]], color = 'black')
                Lines += line([icosaVertices[temp[i][(j-1)%3]],icosaVertices[temp[i][j]]], color = 'black')
        for i in range(13):
            Lines += point(icosaVertices[i], color = icosaColor[i],size = 40)
        show(Lines, axes = False)
        return
    def icosaGenIsolate(self, other):
        end = icosaColor.index(other)
        Lines = point((0,0,0), color = 'white')
        temp = centered[str(self.center)]
        for i in range(1,5):
            if (end in temp[i]):
                for j in range(3):
                    Lines += line([icosaVertices[temp[0]],icosaVertices[temp[i][j]]], color = 'black')
                    Lines += line([icosaVertices[temp[i][(j-1)%3]],icosaVertices[temp[i][j]]], color = 'black')
                break
        for i in range(13):
            Lines += point(icosaVertices[i], color = icosaColor[i],size = 40)
        show(Lines, axes = False)
        return
    def icosaExchange(self,other):
        end = icosaColor.index(other)
        temp = centered[str(self.center)]
        icosaColor[self.center], icosaColor[end] = icosaColor[end], icosaColor[self.center]
        temptwo = []
        for i in range(1,5):
            if (end in temp[i]):
                self.center = end
                for t in temp[i]:
                    if t != end:
                        temptwo.append(t)
                icosaColor[temptwo[0]], icosaColor[temptwo[1]] = icosaColor[temptwo[1]], icosaColor[temptwo[0]]
                return
                

# To examine how the puzzle behaves
I = icosaPuzzle()
counter = ['stop','red','orange','darkkhaki','lawngreen','green','darkseagreen','fuchsia','pink','indigo','cadetblue','cornflowerblue','aquamarine']

@interact
def _(Display = selector(['reduced','all']), push = selector(counter), confirm = selector(['NO','YES']), auto_update = False): 
    if (push == 'stop' and Display == 'all'):
        I.icosaGenTotal()
    elif (push == 'stop' and Display == 'reduced'):
        I.icosaGenLocal()
    elif (push != 'stop'):
        print("Current state:\n")
        if Display == 'reduced':
            I.icosaGenLocal()
        elif Display == 'all':
            I.icosaGenTotal()
        print("These vertices will be affected:\n")
        I.icosaGenIsolate(push)
        if confirm == 'YES':
            I.icosaExchange(push)
            print("Changes reflected:\n")
            if Display == 'reduced':
                I.icosaGenLocal()
            elif Display == 'all':
                I.icosaGenTotal()
                
# solve a scrambled puzzle
#I.icosaPuzzle()
#counter = ['red','orange','darkkhaki','lawngreen','green','darkseagreen','fuchsia','pink','indigo','cadetblue','cornflowerblue','aquamarine']

#@interact
#def _(display = selector(['reduced','all']), locked =selector(['Y','N']), push = selector(counter), auto_update = False):
    #for i in range(10):
        #icosaExchange(counter[randint(0,11)])
    #if display == 'reduced' and lock == 'Y':
        #icosaGenLocal()
    #elif display == 'all' and lock == 'Y':
        #icosGenTotal()
    #elif lock == 'N':
        #icosaExchange(push)
        #if display == 'all':
            #icosaGenTotal()
        #elif display == 'reduced':
            #icosaGenLocal()
```
